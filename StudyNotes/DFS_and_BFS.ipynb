### DFS Tree

```python
def dfs(node, visited):
    if node in visited:
        return
    
    visited.add(node)
    
    dfs(node.left, visited)
    dfs(node.right, visited)
```

```python
# https://leetcode.com/problems/binary-tree-level-order-traversal/
## Binary Tree Level Order Traversal

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        
        queue = [root]
        output = []
        
        while queue:
            new_queue = []
            level_vals = []
            for node in queue:
                if node is not None:
                    level_vals.append(node.val)
                    if node.left:
                        new_queue.append(node.left)
                    if node.right:
                        new_queue.append(node.right)
            
            output.append(level_vals)
            queue = new_queue
        return output
```

```python
## Minimum Genetic Mutation
# https://leetcode.com/problems/minimum-genetic-mutation/
class Solution:
    def minMutation(self, start, end, bank):
        genes = ['A', 'C', 'G', 'T']
        visited = []
        queue = [(start, 0)]
        
        while queue:
            node, step = queue.pop(0)
            if node == end:
                return step
            if node in visited:
                continue
            visited.append(node)
            
            for i in range(len(start)):
                for g in genes:
                    if g != node[i]:
                        new_node = node[:i] + g + node[i+1:]
                        if new_node in bank:
                            queue.append((new_node, step+1))
        return -1
```

```python
# https://leetcode.com/problems/generate-parentheses/
class Solution:
    def generateParenthesis(self, n):
        output = []
        
        def dfs(pos, candid, left, right):
            if pos >= 2*n:
                output.append(candid)
                return
        
            if left + 1 <= n:
                dfs(pos+1, candid + '(', left + 1, right)
            if left >= right + 1:
                dfs(pos+1, candid + ')', left, right+1)
            return
        
        dfs(0, '', 0, 0)
        return output
```

```python
# https://leetcode.com/problems/find-largest-value-in-each-tree-row/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def largestValues(self, root):  
        if not root:
            return []
        
        output = []
        queue = [root]
        
        while queue:
            # iterate every level
            max_val = float('-inf')
            new_queue = []
            for node in queue:
                if node is not None:
                    max_val = max(max_val, node.val)
                    if node.left:
                        new_queue.append(node.left)
                    if node.right:
                        new_queue.append(node.right)
            queue = new_queue
            output.append(max_val)
        return output        
```

```python
# https://leetcode.com/problems/word-ladder/

class Solution:
    def ladderLength(self, beginWord, endWord, wordList):
        if endWord not in wordList:
            return 0
        
        chars = 'abcdefghijklmnopqrstuvwxyz'
        wordList = set(wordList)
        visited = []
        queue = [(beginWord, 1)]
        
        while queue:   
            word, step = queue.pop(0)
            if word == endWord:
                return step
            
            for i in range(len(beginWord)):
                for c in chars:
                    if c != word[i]:
                        new_word = word[:i] + c + word[i+1:]
                        if new_word in wordList and new_word not in visited:
                            visited.append(new_word)
                            queue.append((new_word, step+1))
        return 0
```

```python
# https://leetcode.com/problems/word-ladder-ii/
class Solution:
    def findLadders(self, beginWord, endWord, wordList):
        chars = 'abcdefghijklmnopqrstuvwxyz'
        wordList = set(wordList)
        
        visited = {}
        queue = [[beginWord]]
        
        output = []
        
        while queue:
            chain = queue.pop(0)
            word = chain[-1]
            
            if word == endWord:
                visited[endWord] = len(chain)
                output.append(chain)
                break
                
            if word in visited and len(chain) > visited[word]:
                continue
            visited[word] = len(chain)
            
            for i in range(len(beginWord)):
                for c in chars:
                    if c != word[i]:
                        new_word = word[:i] + c + word[i+1:] 
                        if new_word in wordList and new_word not in visited:
                            queue.append(chain + [new_word])
        
        while queue:
            chain = queue.pop(0)
            word = chain[-1]
            if word == endWord and len(chain) == visited[endWord]:       
                output.append(chain)
        return output      
```
