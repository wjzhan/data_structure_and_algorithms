## Stack

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
        
    def pop(self):
        if len(self.items) < 1:
            return None
        return self.items.pop()
        
    def length(self):
        return len(self.items)
```

```python
# https://leetcode.com/problems/valid-parentheses/
class Solution:
    def isValid(self, s):
        hp = {'(':')',
              '[':']',
              '{':'}'}
        stack = []
        for elem in s:
            if elem in ['(', '[', '{']:
                stack.append(elem)
            else:
                if stack:
                    left = stack.pop()
                    if elem != hp[left]:
                        return False
                else:
                    return False
        
        if not stack:
            return True
        return False
```

```python
# https://leetcode.com/problems/min-stack/
class MinStack:
    
    def __init__(self):
        self.stack = []
        self.mins = []
        self.length = 0
        
    def push(self, val):
        self.stack.append(val)
        if not self.mins:
            self.mins.append(val)
        else:
            prev_min = self.mins[-1]
            if prev_min >= val:
                self.mins.append(val)
            else:
                self.mins.append(prev_min)
        self.length += 1
        
    def pop(self):
        self.length -= 1
        self.stack.pop()
        self.mins.pop()
        
    def top(self):
        return self.stack[-1]
    
    def getMin(self):
        return self.mins[-1]
    
    def length(self):
        return self.length

# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

## Queue

```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def push(self, item):
        self.queue.append(item)
        
    def pop(self):
        if len(self.queue) < 1:
            return None
        return self.queue.pop(0)
        
    def size(self):
        return len(self.queue)
```

```python
#https://leetcode.com/problems/sliding-window-maximum/
class Solution:
    def maxSlidingWindow(self, nums, k):
        output = []
        dqueue = []
        for i in range(len(nums)):
            while dqueue and nums[dqueue[-1]] < nums[i]:
                dqueue.pop()
            dqueue.append(i)
            
            if i >= k-1:
                output.append(nums[dqueue[0]])
            
            if dqueue[0] == i - k + 1:
                dqueue.pop(0)
            
        return output
```

```python
# https://leetcode.com/problems/design-circular-deque

class MyCircularDeque:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the deque to be k.
        """
        self.size = k
        self.deque = []
        

    def insertFront(self, value: int) -> bool:
        """
        Adds an item at the front of Deque. Return true if the operation is successful.
        """
        if len(self.deque) < self.size:
            self.deque.insert(0, value)
            return True
        else:
            return False

    def insertLast(self, value: int) -> bool:
        """
        Adds an item at the rear of Deque. Return true if the operation is successful.
        """
        if len(self.deque) < self.size:
            self.deque.append(value)
            return True
        else:
            return False

    def deleteFront(self) -> bool:
        """
        Deletes an item from the front of Deque. Return true if the operation is successful.
        """
        if self.deque:
            self.deque.pop(0)
            return True
        else:
            return False
        

    def deleteLast(self) -> bool:
        """
        Deletes an item from the rear of Deque. Return true if the operation is successful.
        """
        if self.deque:
            self.deque.pop()
            return True
        else:
            return False
        
        

    def getFront(self) -> int:
        """
        Get the front item from the deque.
        """
        if self.deque:
            return self.deque[0]
        else:
            return -1

    def getRear(self) -> int:
        """
        Get the last item from the deque.
        """
        if self.deque:
            return self.deque[-1]
        else:
            return -1

    def isEmpty(self) -> bool:
        """
        Checks whether the circular deque is empty or not.
        """
        if self.deque:
            return False
        else:
            return True
       
    def isFull(self) -> bool:
        """
        Checks whether the circular deque is full or not.
        """
        if len(self.deque) == self.size:
            return True
        else:
            return False
        


# Your MyCircularDeque object will be instantiated and called as such:
# obj = MyCircularDeque(k)
# param_1 = obj.insertFront(value)
# param_2 = obj.insertLast(value)
# param_3 = obj.deleteFront()
# param_4 = obj.deleteLast()
# param_5 = obj.getFront()
# param_6 = obj.getRear()
# param_7 = obj.isEmpty()
# param_8 = obj.isFull()
```

```python
# https://zhuanlan.zhihu.com/p/125074613
# https://leetcode.com/problems/trapping-rain-water/

class Solution:
    def trap(self, height):
        stacks = [0]
        res = 0
        for i in range(1, len(height)):
            while stacks and height[i] > height[stacks[-1]]:
                idx = stacks.pop()
                if stacks:
                    h = min(height[i], height[stacks[-1]]) - height[idx]
                    res += h * (i - stacks[-1] - 1)
            stacks.append(i)
        return res
```

```python
#https://leetcode.com/problems/basic-calculator
def subcalc(stack):
    output = 0

    if stack[-1] not in ['+', '-']:
        output = stack.pop()

    while stack:
        elem = stack.pop()
        if elem == '-':
            output = output - stack.pop()
        if elem == '+':
            output = output + stack.pop()
    return output


def preprocess(s):
    processed = []
    num = ''
    for char in s:
        if char in ['(', ')', '+', '-']:
            if num:
                processed.append(int(num))
            num = ''
            processed.append(char)
        else:
            num += char
    # process last number
    if num:
        processed.append(int(num))
    return processed


def calculate(s):
    s = s.replace(" ", '')
    # handle corner cases:
    if '+' not in s and '-' not in s:
        s = s.replace('(', '')
        s = s.replace(')', '')
        return int(s)
    # preprocess s for the case of multiple digit
    ss = preprocess(s)

    exp = []
    for elem in ss:
        if elem == ')':
            container = []
            prev = exp.pop()
            while exp and prev != '(':
                container.append(prev)
                prev = exp.pop()
            # process container content to do the calculation
            calc = subcalc(container)
            exp.append(calc)
        else:
            exp.append(elem)
    # now the final exp without any parenthesis
    exp.reverse()
    print(exp)
    result = subcalc(exp)
    return result
            
```
